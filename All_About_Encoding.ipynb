# **Ordinal Encoding**

# **Using loop**

```python
df10.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
attribute=df10.columns
attribute
```

```python
for ord in attribute:
   if is_numeric_dtype(df10[ord]):
     continue
   else:
     ord_encoder=OrdinalEncoder(categories=[df10[ord].unique()])
     enc_values=ord_encoder.fit_transform(df10[[ord]])
     enc_values=pd.DataFrame(enc_values,columns=[ord])
     df10=pd.concat([df10.drop(ord,axis=1),enc_values],axis=1)
df10.head()
```
